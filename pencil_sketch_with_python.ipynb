```python


# Read the original image
image = cv2.imread('masha nd bear (1).jpg')

# Apply Gaussian blur
blurred_image = cv2.GaussianBlur(image, (5, 5), 0)

# Display the blurred image
cv2_imshow(blurred_image)

# Save the blurred image
cv2.imwrite('blurred_image.jpg', blurred_image)
```

```python
inverted_image = cv2.bitwise_not(blurred_image)
```

```python
gray_image = cv2.cvtColor(inverted_image, cv2.COLOR_BGR2GRAY)
```

```python
sketch_image = cv2.divide(255 - gray_image, gray_image, scale=256.0)
```

```python
cv2_imshow(sketch_image)
```

```python
cv2.imwrite('pencil_sketch.jpg', sketch_image)
```
